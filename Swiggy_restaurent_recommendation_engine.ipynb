```python
#pivot table
pv = df.pivot_table(index='Price',columns='Restaurant',values='Avg ratings')
pv.sample(10)
```

```python
df['Restaurant'].value_counts().head(20)
```

```python
dr=pv[["La Pino'Z Pizza",'Nic Natural Ice Creams','Baskin Robbins','Kfc','Subway']]
dr.sample(10)
```

```python
similar_to_analytics=pv.corrwith(dr)
```

```python
corr_analytics=pd.DataFrame(similar_to_analytics,columns=['Correlation'])
corr_analytics.dropna(inplace=True)
corr_analytics.head()
```

```python
corr_analytics.sort_values('Correlation',ascending=False)
```

```python
corr_analytics=corr_analytics.join(de['Avg ratings'])
corr_analytics
```
